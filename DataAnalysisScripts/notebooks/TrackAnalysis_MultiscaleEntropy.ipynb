```python
# Plot the MSE for all the tracks

fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize=(16,9))
for track in tracks:


    # Calculate the MSE entropy of the vertical velocities
    tol = 0.2*np.nanstd(track.Vz)
    
    mean_vel = np.nanmean(track.Vz)
    
    white_noise = np.random.normal(0, np.nanstd(track.Vz), track.trackLen)
    
    mean_vel_noise = mean_vel + white_noise
    
    maxScale_time = 50
    maxScale = int(maxScale_time/track.dT)
    print('Max Scale: {}'.format(maxScale))
    

    mse_z = ent.multiscale_entropy(track.Vz, 3, tol, maxScale)
    
#     mse_whitenoise = ent.multiscale_entropy(white_noise, 3, tol, maxScale)
    
    mse_whitenoise = ent.multiscale_entropy(mean_vel_noise, 3, tol, maxScale)

    scale_factors = np.array(range(maxScale)) + 1

    times_scalefactor = track.dT*scale_factors

    
    axes[0].scatter(track.T, track.Vz, 40, c=next(cycol), alpha = 0.5, label = track.trackName)
    axes[0].plot(track.T, mean_vel_noise, c=next(cycol), alpha = 0.5, label = 'mean velocity + White noise')
    axes[0].set_xlabel('Time (s)')
    axes[0].set_ylabel('Vertical velocity')
    axes[0].set_title('Velocity time trace')
    axes[0].legend()
    
    
    axes[1].scatter(times_scalefactor, mse_z, 40, c=next(cycol), alpha = 0.5, label = track.trackName)
    axes[1].plot(times_scalefactor, mse_whitenoise, 'k-', alpha = 0.5, label = 'White noise')
    axes[1].vlines(track.dT*10, 0, 2.5, color=next(cycol))
    axes[1].set_xscale('log')
    axes[1].set_xlabel('Scale factor')
    axes[1].set_ylabel('Sample Entropy')
    axes[1].set_title('Multi-scale Entropy Analysis')
    plt.legend()
```
