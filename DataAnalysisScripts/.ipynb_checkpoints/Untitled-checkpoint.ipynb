# STFT 

```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

fs = 10e3
N = 1e5
amp = 2 * np.sqrt(2)
noise_power = 0.01 * fs / 2
time = np.arange(N) / float(fs)
mod = 500*np.cos(2*np.pi*0.25*time)
carrier = amp * np.sin(2*np.pi*3e3*time + mod)
noise = np.random.normal(scale=np.sqrt(noise_power), size=time.shape)
noise *= np.exp(-time/5)
x = carrier + noise


plt.figure(figsize=(4,3),dpi=150)
plt.plot(time,x)
plt.title('Signal')
plt.ylabel('Signal')
plt.xlabel('Time [sec]')
plt.show()

# Plot the frequency spectrum
plt.figure(figsize=(4,3),dpi=150)
f, t, Zxx = signal.stft(x, fs, nperseg=1000)
plt.pcolormesh(t, f, np.abs(Zxx), vmin=0, vmax=amp)
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

# CWT

```python
time = np.linspace(-1, 1, 200, endpoint=False)
freq = 1/(time[1]-time[0])
print(50*'*')
print('Sampling frequency {} Hz'.format(freq))
print(50*'*')
sig  = np.cos(2 * np.pi * 7 * time) + signal.gausspulse(time - 0.4, fc=2)
widths = np.arange(1, 31)
cwtmatr = signal.cwt(sig, signal.ricker, widths)


plt.figure(figsize=(4,3),dpi=150)
plt.plot(time,sig)
plt.title('Signal')
plt.ylabel('Signal')
plt.xlabel('Time [sec]')
plt.show()

# Plot the frequency spectrum
plt.figure(figsize=(4,3),dpi=150)

plt.imshow(cwtmatr, extent=[-1, 1, 1, 31], cmap='PRGn', aspect='auto',vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())
plt.title('CWT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
