# Script to test the MSD calculation and Least-squares fitting to Taylor's equation

## Generate test data for tumblers to validate the MSD calculation

```python
import GravityMachineTrack
import imp
import msdanalyzer
import pandas as pd
import numpy as np
import os
imp.reload(GravityMachineTrack)
imp.reload(msdanalyzer)
from lmfit import minimize, Parameters
from lmfit.models import ExpressionModel



import matplotlib.pyplot as plt
%matplotlib notebook

def Taylor_Function(x, v, tau):
    
    return 2*(v**2)*tau*(x - tau*(1 - np.exp(-x/tau)))

v_sim = 1
tau_sim = 5

msd1 = msdanalyzer.msdanalyzer(testFlag=True,v = v_sim, tau=tau_sim)
msd1.computeMSD()


```

```python
# Plot the MSD
msd1.plotMSD(figname = 1, plot_analytical=False)
```

## Comparison to analytical solution for tumblers

```python
#-----------------------------------------------------------
# Compare the calculated MSD against the analytical solution
#-----------------------------------------------------------

x = msd1.delaysCommon[1:] - msd1.delaysCommon[1]
y = msd1.MSD_3D[1:] - msd1.MSD_3D[1] 
y_err  = msd1.stdev_3D[1:]
y_MSD = msd1.MSD_3D[1:] - msd1.MSD_3D[1] 

fig, (ax1, ax2) = plt.subplots(figsize=(8,4),nrows=1, ncols=2)

ax1.plot(x, y_MSD , 'ro', label='calculated MSD for tumblers (3D)', alpha = 0.5)
ax1.plot(x, Taylor_Function(x=x,v=v_sim, tau=tau_sim) , 'k--', label='Analytical solution for tumblers')

plt.legend(loc='best')



ax2.plot(x, y_MSD , 'ro', label='calculated MSD for tumblers (3D)', alpha = 0.5)
ax2.plot(x, Taylor_Function(x=x,v=v_sim, tau=tau_sim) , 'k--', label='Analytical solution for tumblers')

plt.legend(loc='best')

ax1.set_yscale('log')
ax1.set_xscale('log')
plt.show()
```

## Non-linear least-squares fitting of MSD data

```python
# Expression based model for non-linear least-squares fitting
from numpy import exp

gmod = ExpressionModel('2*(v**2)*tau*(x - tau*(1 - exp(-x/tau)))')
result = gmod.fit(y, x=x, tau=1, v=2, nan_policy='omit')
```

```python
# Plot the results
print(result.fit_report())

fig, (ax1, ax2) = plt.subplots(figsize=(8,4),nrows=1, ncols=2)


ax1.plot(x, y, 'bo', label = 'MSD')
ax1.plot(x, Taylor_Function(x=x,v=v_sim, tau=tau_sim), 'r.', alpha = 0.2, label='Analytical solution')

ax1.plot(x, result.init_fit, 'k--', label='initial fit')
ax1.plot(x, result.best_fit, 'k-', label='best fit')

ax2.plot(x, y, 'bo',label = 'MSD')
ax2.plot(x, Taylor_Function(x=x,v=v_sim, tau=tau_sim), 'r.', alpha = 0.2, label='Analytical solution')

ax2.plot(x, result.init_fit, 'k--', label='initial fit')
ax2.plot(x, result.best_fit, 'k-', label='best fit')
ax2.set_yscale('log')
ax2.set_xscale('log')

plt.legend(loc='best')
plt.show()
```

```python
# Diagnostic tests
fig, (ax1, ax2) = plt.subplots(figsize=(8,4),nrows=1, ncols=2)

ax1.plot(x, y_err , 'ro', label='Stddev in squared displacement', alpha = 0.5)

plt.legend(loc='best')



ax2.plot(x, y_err , 'ro', label='Stddev in squared displacement', alpha = 0.5)

plt.legend(loc='best')

ax1.set_yscale('log')
ax1.set_xscale('log')
plt.show()
```
