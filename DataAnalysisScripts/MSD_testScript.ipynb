```python
# Setup a non-linear least-squares minimization problem
from lmfit.models import ExpressionModel
from numpy import exp

def residual(params, x, data, eps_data):

    tau = params['tau']
    v = params['v']
    
    # Taylor function
    model = 2*(v**2)*tau*(x - tau*(1 - np.exp(-x/tau)))

    return (data - model)/eps_data

x = msd1.delaysCommon[1:] - msd1.delaysCommon[1]
y = msd1.MSD_Z[1:] - msd1.MSD_Z[1] 
y_err  = msd1.stdev_Z[1:]

params = Parameters()

params.add('tau', value=2, min=0.0)
params.add('v', value=10, min = 0.0)

out = minimize(residual, params, args=(x, y, y_err))

print(out)

tau_fit = out.params['tau']
v_fit = out.params['v']
```
